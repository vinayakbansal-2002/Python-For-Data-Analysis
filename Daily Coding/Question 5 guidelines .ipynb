```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("iris.data",header=None,names=['sepal_length','sepal_width','petal_length','petal_width','class'])
df
```

### a. Plot bar chart to show the frequency of each class label in the data.

```python
df['class'].value_counts().plot.bar()
```

### Draw a scatter plot for Petal width vs sepal width.

```python
plt.scatter(df['petal_width'],df['sepal_width'])
```

### c. Plot density distribution for feature petal length.

```python
df['petal_length'].plot.density()
```

### d. Use a pair plot to show pairwise bivariate distribution in the Iris Dataset.

```python
import seaborn as sns
```

```python
sns.pairplot(df)
```
